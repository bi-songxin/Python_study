```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
# numpy中的主要处理基于C/C++实现
# np.array()方法接收py列表作为参数，生成numpy数组（numpy.ndarray）
x = np.array([1.0, 2.0, 3.0])
print(x)
type(x)
```

```python
# numpy算术运算
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

```python
# 广播
x = np.array([1.0, 2.0, 3.0])
# 2称为标量（单一数值）
x / 2
```

```python
# numpy的N维数组
A = np.array([[1, 2], [3, 4]])
print(A)
# 2 x 2的矩阵A
print(A.shape)
# 矩阵的数据类型
print(A.dtype)
# 矩阵的广播
print(A * 10)
```

数学上：
- 一维数组称为向量
- 二维数组称为矩阵
- 一般化的向量或矩阵称为张量

```python
# 矩阵的广播
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
A * B
```

```python
# numpy访问元素
X = np.array([[51, 13], [12, 45], [42, 61]])
print(X)
# 将X转换为一维数组(永久改变)
X = X.flatten()
print(X)
# 获取索引为0、2、4的元素
print(X[np.array([0, 2, 4])])
print(X > 50)
print(X[X > 50])
```

```python
# 生成数据
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 画图
# 绘制图形
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='sin')
# 轴标签
plt.xlabel('x')
plt.ylabel('y')
# 标题
plt.title('sin & cos')
# 线解释
plt.legend()
# 展示图形
plt.show()
```
